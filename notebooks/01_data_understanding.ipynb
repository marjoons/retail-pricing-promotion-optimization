```python
def dataset_overview(
    name: str,
    dataframe: pd.DataFrame,
    sample_type: str = "Complete",
) -> dict:
    """Return high-level information about a dataset."""

    memory_mb = (
        dataframe.memory_usage(deep=True).sum()
        / 1024**2
    )

    return {
        "dataset": name,
        "sample_type": sample_type,
        "rows_loaded": dataframe.shape[0],
        "columns": dataframe.shape[1],
        "memory_mb": round(memory_mb, 2),
        "duplicate_rows": int(
            dataframe.duplicated().sum()
        ),
    }


overview = pd.DataFrame(
    [
        dataset_overview(
            "M5 Calendar",
            m5_calendar,
        ),
        dataset_overview(
            "M5 Selling Prices",
            m5_prices_sample,
            "100,000-row sample",
        ),
        dataset_overview(
            "M5 Historical Sales",
            m5_sales_sample,
            "100-row sample",
        ),
        dataset_overview(
            "Synthetic Retail Pricing",
            synthetic_retail,
        ),
        dataset_overview(
            "Synthetic Market Notes",
            market_notes,
        ),
    ]
)

overview
```

```python
def missing_value_summary(
    dataframe: pd.DataFrame,
) -> pd.DataFrame:
    """Summarize columns containing missing values."""

    missing_count = dataframe.isna().sum()

    summary = pd.DataFrame(
        {
            "missing_count": missing_count,
            "missing_percentage": (
                missing_count
                / len(dataframe)
                * 100
            ).round(2),
        }
    )

    return (
        summary[summary["missing_count"] > 0]
        .sort_values(
            "missing_percentage",
            ascending=False,
        )
    )


ffor name, dataframe in datasets.items():
    columns = dataframe.columns.tolist()

    print(f"\n{name}")
    print("-" * len(name))
    print(f"Total columns: {len(columns)}")
    print(f"First 15 columns: {columns[:15]}")

    if len(columns) > 15:
        print(f"Last 5 columns: {columns[-5:]}")
```
